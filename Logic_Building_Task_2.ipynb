**Unique Words in a Sentence**

```python
def unique_words(sentence):
  words = list(sentence.split())
  unique_words = set(words)
  return unique_words

sentence = "python is easy and python is powerful"
result = unique_words(sentence)
print("Unique words count:", len(result))
print("Unique words:", result)
```

**Highest Salary from Employee Data**

```python
def highest_salary_emp(employees):
  highest_salary = 0
  highest_earner_name = None
  for name, salary in employees.items():
    if salary > highest_salary:
      highest_salary = salary
      highest_earner_name = name
  return highest_earner_name, highest_salary

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
name, salary = highest_salary_emp(employees)
print(f"Highest Salary: {name} - {salary}")
```

**3. Find Maximum and Minimum Values**

```python
def find_max_min(numbers):
  max = numbers[0]
  min = numbers[0]
  for num in numbers:
    if num > max:
      max = num
    elif num < min:
      min = num
  return max, min

numbers = [45, 22, 89, 10, 66]
print(f"List: {numbers}")
max, min = find_max_min(numbers)
print(f"Max: {max} \nMin: {min}")
```

**4. Count Products Above a Price Threshold**

```python
def Prods_abv_thrshold(prices, threshold):
  count = 0
  for price in prices:
    if price > threshold:
      count += 1
  return count

prices = [450, 1200, 899, 1500, 300]
threshold = 1000
print(f"Prices above {threshold}: {Prods_abv_thrshold(prices, threshold)}")
```

**5. Calculate Attendance Percentage**

```python
def attendance_percent(attendance):
  present_count = 0
  for i in attendance:
    if i == 'P':
      present_count += 1

  attendance_percent = (present_count / len(attendance)) * 100
  return attendance_percent

attendance = ["P", "P", "A", "P", "P"]
print("Attendance Percentage:", attendance_percent(attendance))
```

**6. Remove Duplicate Phone Numbers**

```python
def remove_duplicates(phone_numbers):
  unique_numbers = set(phone_numbers)
  return unique_numbers

phone_numbers = [9876543210, 9123456789, 9876543210]
print("Unique phone numbers:", remove_duplicates(phone_numbers))
```

**7. Count Character Frequency**

```python
def char_feq(string):
  freq = {} # Initialize as an empty dictionary instance
  for char in string:
    if char in freq:
      freq[char] += 1
    else:
      freq[char] = 1 # Add character if not present
  return freq

string = "pythonp"
print(char_feq(string))
```

**8. Convert List to Tuple**

**9. Check if a Key Exists in a Dictionary**

```python
def key_in_dict(dict, key):
  if key in dict.keys():
    print(f"{key} exists")
  else:
    print(f"{key} does not exist")

dict = {"Employee": "Ravi", "Salary": 75000}
key = "Employee"
key_in_dict(dict, key)
```

**10. Calculate Average Marks**
