**1. User Login Check**

```python
def user_login_check(input_username, input_password):
  '''
  Checks username and password for valid credentials
  '''
  username = "admin"
  password = "1234"

  if input_username == username and input_password == password:
    return "Login Successful"
  else:
    return "Invalid Credentials"


result = user_login_check("admin", "1234")
print(result)
```

**2. pass / Fail Analyzer**

```python
def pass_fail_analyzer(marks):
  '''
  Analyzes marks and returns pass/fail status
  '''

  pass_count = 0
  fail_count = 0

  for mark in marks:
    if mark >= 50:
      pass_count += 1
    else:
      fail_count += 1

  return pass_count, fail_count

marks = [45, 78, 90, 33, 60]
pass_count, fail_count = pass_fail_analyzer(marks)
print("Pass Count:", pass_count)
print("Fail Count:", fail_count)

```

**3. Simple Data Cleaner**

```python
def data_cleaner(names):
  '''
  Clean and standardize usernames
  '''
  cleaned_data = []
  for name in names:
    cleaned_item = name.strip().lower()
    cleaned_data.append(cleaned_item)

  return cleaned_data

names = [" Alice ", "bob", " CHARLIE "]
print(data_cleaner(names))
```

**4. Message Length Analyzer**

```python
def message_length_analyzer(messages):
  '''
  Analyzes message lengths
  '''
  results = []
  for message in messages:
    length = len(message)
    flag = length > 10
    results.append({
        "message" : message,
        "length" : length,
        "flag" : flag})

  return results

messages = ["Hi", "Welcome to the platform", "OK"]
analysis = message_length_analyzer(messages)

for result in analysis:
  print(f"Message: '{result['message']}' | Length: {result['length']}")
  if result['flag']:
    print("Flag: 'Message is longer than 10 characters.'")
  print()

```

**5. Error Message Detector**

```python
def error_message_detector(logs):
  '''
  Detects number of errors in system logs
  '''
  error_count = 0
  for log in logs:
    if "ERROR" in log:
      error_count += 1

  return error_count

logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = error_message_detector(logs)
print("Number of Errors:", error_count)
```
